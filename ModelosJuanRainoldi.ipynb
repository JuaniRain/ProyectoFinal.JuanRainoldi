#Maching Learning

Se realizará el entrenamiento de varios modelos para ver cual es el que presenta mejores respuestas para el problema planteado.

```python
#Se importan las librerías
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn import neighbors
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

Se separan los datos  'x' (atributos), 'y' (etiquetas) e 'xx' (auxiliar para el important feature). A estos datos se lo identificará como **Data Frame Normal**

```python
#Se separan en valores de x e y
x=creditos.drop(['Credit_Mix'],axis=1)
y=creditos['Credit_Mix']
xx=x.copy()
```

```python
#Se importan librerías
from sklearn.model_selection import cross_val_score, KFold
from numpy import std
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score
```

#Se procede a entrenar los modelos:

**RandomForest** con **Data Frame Normal**

```python
#Se configura el KFold con 6 splits
kf = KFold(n_splits=6, shuffle=True, random_state=42)

#Se crea una  lista para almacenar los resultados
Recallrf = []
Accuracyrf=[]

#Se convierte x e y a matrices NumPy
x = np.array(x)
y = np.array(y)

#Se itera a través de cada fold
for train_index, test_index in kf.split(x):
    #Se dividen los datos en entrenamiento y prueba
    X1_train, X1_test = x[train_index], x[test_index]
    y1_train, y1_test = y[train_index], y[test_index]

    #Se entrena el modelo
    modelo = RandomForestClassifier(max_depth=5, random_state=42)
    modelo.fit(X1_train, y1_train)

    #Se realiza las predicciones
    y_pred = modelo.predict(X1_test)

    #Se calcula el Recall para cada interación y se carga a la lista
    rec=recall_score(y1_test, y_pred, average='macro')
    Recallrf.append(rec)

    #Se calcula el Accuracy para cada interación y se carga a la lista
    ac=accuracy_score(y1_test, y_pred)
    Accuracyrf.append(ac)

    #Se muestra el resultado del Recall de cada iteración
    print(f'El recall en esta iteración: {rec}')
    #Se muestra el resultado del Accuracy de cada iteración
    print(f'El accuracy en esta iteración: {ac}')

#Se muestra el Recall promedio de le las iteraciones y el desvío standard
print()
print('El Recall promedio:', np.mean(Recallrf))
stand=std(Recallrf,axis=0)
print(f'Desvio standard: {stand}')

#Se muestra el Accuracy promedio de le las iteraciones y el desvío standard
print()
print('El Accuracy promedio:', np.mean(Accuracyrf))
stand=std(Accuracyrf,axis=0)
print(f'Desvio standard: {stand}')
```

Acorde al valor de las métricas obtenidas, se puede afirmar que el modelo sirve para resolver el problema planteado.
